````python
import sys
from pathlib import Path

# Agregar raíz del proyecto al path para importar src
PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from src.config import PROCESSED_DIR, ACTIVE_PROFILE

plt.style.use("seaborn-v0_8-whitegrid")
plt.rcParams["figure.figsize"] = (12, 5)
````

## 1. Cargar dataset procesado

````python
# Cargar dataset procesado
data_path = PROCESSED_DIR / "unemployment_processed.csv"

df = pd.read_csv(data_path, parse_dates=["date"])
print(f"Dataset: {len(df)} filas × {len(df.columns)} columnas")
print(f"Rango temporal: {df['date'].min()} — {df['date'].max()}")
df.head(10)
````

## 2. Estadísticas descriptivas

````python
df.describe()
````

## 3. Serie temporal de desempleo

````python
fig, ax = plt.subplots()

ax.plot(df["date"], df["unemployment_rate"], linewidth=1.5, color="#2c3e50")
ax.set_title("Tasa de Desempleo — Serie Temporal", fontsize=14, fontweight="bold")
ax.set_xlabel("Fecha")
ax.set_ylabel("Tasa de desempleo (%)")
ax.axhline(
    y=df["unemployment_rate"].mean(),
    color="#e74c3c",
    linestyle="--",
    linewidth=1,
    label=f"Media: {df['unemployment_rate'].mean():.1f}%",
)
ax.legend()
fig.tight_layout()
plt.show()
````
